```r
library(igraph)
library(dplyr)
library(microbenchmark)
library(here)
```

```r
nodes <- c(10,32, 100, 316, 1000, 3162, 10000)
setwd(here::here()) 
setwd('..')
```

```r
el <- read.csv(paste(getwd(),'/Data/G_ ',as.character(10),' _p1',sep=''))
el <- el[,2:3]
G <- graph_from_data_frame(el)

p1_comp <- microbenchmark("Components" = {components(G)})
p1_comp <- summary(p1_comp)

p1_pr <- microbenchmark("PageRank" = {page_rank(G, damping=0.85)})
p1_pr <- summary(p1_pr)
```

```r
p1_comp
```

```r
rm(p1_comp,p1_pr)
```

```r
page_rank(G,damping=0.85)
```

```r
set.seed(1)
for (n in nodes){
    
    el <- read.csv(paste(getwd(),'/Data/G_ ',as.character(n),' _p1',sep=''))
    el <- el[,2:3]
    G <- graph_from_data_frame(el)
    
    
    a <- microbenchmark(n = {components(G)})
    a <- summary(a)

    b <- microbenchmark(n = {page_rank(G, damping=0.85)})
    b <- summary(b)

    if (n==10){
        p1_comp <- a
        p1_pr <- b
    }else{
        p1_comp <- rbind(p1_comp, a)
        p1_pr <- rbind(p1_pr,b)
    }
}
```

```r
set.seed(1)
for (n in nodes){
    
    el <- read.csv(paste(getwd(),'/Data/G_ ',as.character(n),' _p2',sep=''))
    el <- el[,2:3]
    G <- graph_from_data_frame(el)
    
    
    a <- microbenchmark(n = {components(G)})
    a <- summary(a)

    b <- microbenchmark(n = {page_rank(G, damping=0.85)})
    b <- summary(b)

    if (n==10){
        p2_comp <- a
        p2_pr <- b
    }else{
        p2_comp <- rbind(p2_comp, a)
        p2_pr <- rbind(p2_pr,b)
    }
}
```

```r
p2_comp[7,2:7] <- p2_comp[7,2:7]*1000
```

```r
p2_pr
```

```r
p2_pr[6:7,2:7] <- p2_pr[6:7,2:7]*1000
```

```r
p2_pr
```

```r
p1_comp_time <- p1_comp$median
p2_comp_time <- p2_comp$median
p1_pr_time <- p1_pr$median
p2_pr_time <- p2_pr$median

rtimes_df <- data.frame(p1_comp_time,p2_comp_time,p1_pr_time,p2_pr_time)
```
